```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (change 'your_dataset.csv' to your actual file path)
df = pd.read_csv("../Dataset_with_Features/dataset_10000.csv")

# Shuffle the rows of the dataset
df_shuffled = df.sample(frac=1).reset_index(drop=True)

# Save the shuffled dataset to a new CSV file
df_shuffled.to_csv("dataset_10000.csv", index=False)

# Check and print the number of 'good' and 'bad' URLs
num_good = df_shuffled[df_shuffled['label'] == 'good'].shape[0]
num_bad = df_shuffled[df_shuffled['label'] == 'bad'].shape[0]
print(f"Number of good URLs: {num_good}")
print(f"Number of bad URLs: {num_bad}")

# Display the head of the dataframe
print("\nHead of the shuffled dataframe:")
print(df_shuffled.head())

# Draw a histogram
labels = ['legitimate', 'phishing']
counts = [num_good, num_bad]

plt.bar(labels, counts, color=['green', 'red'])
plt.title('Histogram of Legitimate and Phishing URLs')
plt.xlabel('URL Label')
plt.ylabel('Count')
plt.show()
```
